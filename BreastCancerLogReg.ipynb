```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings("ignore")
```

# Data Loading

```python
breast_cancer_dateset = sklearn.datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(breast_cancer_dateset.data, columns=breast_cancer_dateset.feature_names)
```

# Data Exploration and Preprocessing

```python
df.head(5)
```

```python
# adding label column to df
df['label'] = breast_cancer_dateset.target
```

```python
df.shape
```

```python
# check for missing values
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['label'].value_counts()
# 1 = Benign
# 2 = Malignant
```

```python
df.groupby('label').mean()
```

# Train/Test Split

```python
x = df.drop(columns='label', axis=1)
y = df['label']
```

```python
# splitting data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x.shape, x_train.shape, x_test.shape
```

# Modeling

```python
# model training logistic regression (binary classification problem)
model = LogisticRegression(max_iter=2000)
```

```python
model.fit(x_train, y_train)
```

# Model Evaluation

```python
# accuracy on training data
x_train_pred = model.predict(x_train)
train_data_acc = accuracy_score(y_train, x_train_pred)
```

```python
print(f'Accuracy on training data = {train_data_acc}')
```

```python
# accuracy on testing data
x_test_pred = model.predict(x_test)
test_data_acc = accuracy_score(y_test, x_test_pred)
```

```python
print(f'Accuracy on testing data = {test_data_acc}')
```

```python
model.score(x_train, y_train), model.score(x_test, y_test)
```

# Example Prediction

```python
def predict_tumor(model, input_data=None):
    """
    Predict tumor type using the trained model.

    Parameters:
        model: trained ML model with .predict()
        input_data (tuple, optional):
            Feature values. Defaults to example data.

    Returns:
        prediction (int): model output (0 or 1)
    """

    # Default data if none provided
    if input_data is None:
        input_data = (
            18.12, 10.21, 121.9, 995.4, 0.1168,
            0.2812, 0.2947, 0.1493, 0.2386, 0.07902,
            1.082, 0.9186, 8.44, 150.9,
            0.00651, 0.0482, 0.0551, 0.0162, 0.0294, 0.00605,
            25.11, 17.61, 182.3, 1987.6,
            0.1598, 0.6721, 0.7043, 0.2689, 0.4527, 0.1204
        )

    input_data = np.asarray(input_data).reshape(1, -1)
    prediction = model.predict(input_data)[0]

    label = "Malignant" if prediction == 0 else "Benign"
    print(f"Prediction: {prediction} → {label}")

    return prediction
```

```python
predict_tumor(model)
```
